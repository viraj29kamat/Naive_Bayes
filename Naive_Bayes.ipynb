```python
salary_test = pd.read_csv(r"C:\Users\VIRAJ KAMAT\Desktop\Data Science\Assignments\Naive bayes\SalaryData_Test.csv")
salary_test
```

```python
salary_train.isnull().sum()
```

```python
salary_test.isnull().sum()
```

```python
train = salary_train.iloc[:,0:13]

train = pd.get_dummies(train)
```

```python
finaltrain = pd.concat([train, salary_train['Salary']],axis=1)
```

```python
test = salary_test.iloc[:,0:13]

test = pd.get_dummies(test)
```

```python
finaltest = pd.concat([test, salary_test['Salary']],axis=1)
```

```python
finaltrain_x = finaltrain.values[:,0:102]
finaltrain_y = finaltrain.values[:,102]
```

```python
finaltest_x = finaltest.values[:,0:102]
finaltest_y = finaltest.values[:,102]
```

## Preparing a naive bayes model

```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB as MB
from sklearn.naive_bayes import GaussianNB as GB
```

## Multinomial Naive Bayes

```python
classifier_mb = MB()
classifier_mb.fit(finaltrain_x,finaltrain_y)
train_pred_m = classifier_mb.predict(finaltrain_x)
accuracy_train_m = np.mean(train_pred_m==finaltrain_y)

test_pred_m = classifier_mb.predict(finaltest_x)
accuracy_test_m = np.mean(test_pred_m==finaltest_y)


print('Training accuracy is:',accuracy_train_m,'\n','Testing accuracy is:',accuracy_test_m)
```

 ## Gaussian Naive Bayes

```python
classifier_gb = GB()
classifier_gb.fit(finaltrain_x,finaltrain_y) 
train_pred_g = classifier_gb.predict(finaltrain_x)
accuracy_train_g = np.mean(train_pred_g==finaltrain_y)

test_pred_g = classifier_gb.predict(finaltest_x)
accuracy_test_g = np.mean(test_pred_g==finaltest_y)

print('Training accuracy is:',accuracy_train_g,'\n','Testing accuracy is:',accuracy_test_g)
```

## Gaussian Naive Bayes gives higher Accuracy : 80%
